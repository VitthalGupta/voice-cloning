```python
from TTS.tts.models.bark import Bark
from TTS.tts.configs.bark_config import BarkConfig
from scipy.io.wavfile import write as write_wave
import os
```

```python
config = BarkConfig()
model = Bark.init_from_config(config)
model.load_checkpoint(
    config, checkpoint_dir="../bark/", eval=True)
# model.to("metal")
```

```python
# cloning a speaker.
# It assumes that you have a speaker file in `bark_voices/speaker_n/speaker.wav` or `bark_voices/speaker_n/speaker.npz`
# print(os.listdir())

text = """
Hello, my name is brian greene, and this is my AI voice clone, how are you doing? [laugh]
"""

output_dict = model.synthesize(
    text, config, speaker_id="processed", voice_dirs="../data/")
```

```python
write_wave("output.wav", 24000, output_dict["wav"])
```
